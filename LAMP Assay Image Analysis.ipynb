## Logistic Regression
Using LR with most frequent color

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
def most_frequent_colour(image):

    w, h = image.size
    pixels = image.getcolors(w * h)

    most_frequent_pixel = pixels[0]

    for count, colour in pixels:
        if count > most_frequent_pixel[0]:
            most_frequent_pixel = (count, colour)

    return most_frequent_pixel
```

```python
freq = []
for image in image_list:
    h = most_frequent_colour(image)
    freq.append(h[1])
freq
```

# Color Normalization 

Uses the bottom-right corner of the image as a reference point. It then uses the differences in the average values of the corner pixels and perfect white and applies it to the entire image. 

```python
im = Image.open("./CITAS/colornormtest/test4.jpg")
im2 = Image.new(im.mode, im.size)

width, height = im.size

avgarray = []

avg = [0, 0, 0]

for i in range(0,3):
    pxcount = 0
    for x in range((width // 4) * 3, width):
        for y in range((height // 4) * 3, height):
            avg[i] = avg[i] + im.getpixel( (x,y) )[i]
            pxcount = pxcount + 1
    avg[i] = avg[i] // pxcount    

avg

diff = []
diff.append(255 - avg[0])
diff.append(255 - avg[1])
diff.append(255 - avg[2])
diff
```

```python
for x in range(0, width):
    for y in range(0, height):
        current_color = im.getpixel( (x,y) )
        new = []
        for i in range(0,3):
            if ((current_color[i] + diff[i]) <= 255):
                new.append(current_color[i] + diff[i])
            else:
                new.append(255)
        new_color = (new[0], new[1], new[2])
        im2.putpixel( (x,y), new_color)
        
im2.show()
```

## CROP
Gets the center box of an imaginary 3x3 grid on the image 

```python
left = (width // 3)
up = (height // 3)

im3 = im2.crop(box=(left, up, left*2, up*2))
im3.show()
```

## TURN ALL B&W COLORS TO BLACK
Using standard deviation, make all pixels that are black to white all black. 

```python
im4 = Image.new(im3.mode, im3.size)

width3, height3 = im3.size

bw_tolerance = 18

for x in range(0, width3):
    for y in range(0, height3):
        current_color = im3.getpixel( (x,y) )
        curcolorlist = [current_color[0], current_color[1], current_color[2]]
        new = []
        for i in range(0,3):
            if (numpy.std(curcolorlist) >= bw_tolerance):
                new.append(current_color[i])
            else:
                new.append(0)
        new_color = (new[0], new[1], new[2])
        im4.putpixel( (x,y), new_color)

im4.show()
```
